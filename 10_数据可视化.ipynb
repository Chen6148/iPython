## Pandas可视化
>一图胜千言，Pandas的可视化图表可以让数据通过图表的形式展示数据的比例、分布、趋势等信息，可以让数据自己说话。

Pandas提供了`df.plot()`函数，可以快速方便的将Series和DataFrame类型的数据进行图表绘制；plot函数主要是`matplotlib.axes.Axes.plot`的封装。

接下来，将详细介绍plot函数以及常用的可视化图表。


## plot()函数参数详解
**语法格式**：`df.plot(x=None, y=None, kind='line',subplots=False, figsize=None, title=None, grid=None, legend=True, style=None, color=None, **kwds)`

**常用参数**:
- `x`、`y`：表示x轴或y轴的标签或位置，指定显示的索引，默认为None；
- `title`：设置图表标题；
- `fontsize`: 设置轴刻度的字体大小，单位是pt(磅)； 
- `figsize`：图形尺寸大小，(width, height)
- `grid`：图形是否有网格，默认None legend：子图的图例
- `style`：指定折线图中线条的类型，list or dict；
- `color`：指定折线图中线条的颜色，list or dict；
- `legend`：设置图例
- `kind`：表示绘图的类型，默认为line（折线图）；
    - line：折线图
    - bar/barh：柱状图，或条形图（纵向/横向）
    - pie：饼状图
    - hist：直方图（数值频率分布）
    - box：箱型图
    - scatter：散点图    
    ...
- `subplots`：对数据表中的每一列分别作子图，默认False；


**辅助参数**
```python
ax = df.plot()
ax.set_title()      # 设置标题
ax.set_ylabel('X')  # 设置x轴标签
ax.set_xlabel('Y')  # 设置y轴标签
```

### 设置图表数据-x轴和y轴数据设置
参数x、y的使用，常用在折线图、柱状图、面积图、散点图等图表上，指定x轴和y轴上的数据；

- 若是Series数据，则无须传入x、y值，默认x = 索引；    
- 若是DataFrame数据，则`x`、`y`参数可指定对应轴上的数据；

````python
# 创建数据
data = np.random.randint(60, 100, [31, 3])
date_index = pd.date_range('2023-01-01', periods=31)
df = pd.DataFrame(data=data,index=date_index,
    columns=['Chinese', 'Math', 'English'])
df.reset_index(inplace=True)  # 重置索引列
display(df.head())

# 设置x轴和y轴数据
df.plot(x='index', y='English')  # 设置标题
````

### 设置图表样式-线条样式（类型和颜色）
在`df.plot(style=None, color=None)`中，通过向style和color两个参数传入列表，可以对多个曲线依此进行样式设置。

**线条样式**： 

|标识符|样式|标识符|样式|标识符|样式|
|:---:|:---:|:---:|:---:|:---:|:---:|
|`:`|虚线|`-.`|虚实相间|`--`|长虚线|`-`|实线|
|`.`|点|`*-`|实线，数值为★|`^-`|实线，数值为▲|||

````python
# 创建数据
data = np.random.randint(60, 100, [8, 3])
date_index = ['01','02','03','04','05','06','07','08']
df = pd.DataFrame(data=data,index=date_index,
    columns=['语文', '数学', '英语'])

# 设置线条样式和颜色
ax = df.plot(style=['-','--','.-'],
                color=['r','g','b'])

ax.set_title('语数外考试月成绩', fontsize=15)
ax.set_xlabel('考试月',fontsize=12)
ax.set_ylabel('得分',fontsize=12)
````

### 设置图表样式-图例

`df.plot(legend=True)`中的legend参数，默认值为True，即显示图例；
- False：隐藏图例
- reverse：倒排图例

````python
# 创建数据
data = np.random.randint(60, 100, [8, 3])
date_index = ['01','02','03','04','05','06','07','08']
df = pd.DataFrame(data=data,index=date_index,
    columns=['语文', '数学', '英语'])

# 不显示图例
ax = df.plot(legend=False)

ax.set_title('语数外考试月成绩', fontsize=15)
ax.set_xlabel('考试月',fontsize=12)
ax.set_ylabel('得分',fontsize=12)
````

### 设置图表样式-图表类型
1. `df.plot()`的kind参数，可以指定图表的类型：
- line：折线图，默认
- pie：饼图
- bar：柱状图
- barh：横向柱状图
- hist：直方图
- kde、density：核密度估计图
- box：箱形图
- area：面积图
- scatter：散点图
- hexbin：六边形分箱图
2.使用点运算符绘制图表类型
- `df.plot.pie()`：饼图
- `df.plot.bar()`：柱状图
- `df.plot.hist()`：直方图

````python
# 创建数据
data = np.random.randint(60, 100, [8, 3])
date_index = ['01','02','03','04','05','06','07','08']
df = pd.DataFrame(data=data,index=date_index,
    columns=['语文', '数学', '英语'])

# 使用kind参数绘制柱状图
ax = df.plot(kind='bar')

ax.set_title('语数外考试月成绩', fontsize=15)
ax.set_xlabel('考试月',fontsize=12)
ax.set_ylabel('得分',fontsize=12)
````

````python
# 创建数据
data = np.random.randint(60, 100, [8, 3])
date_index = ['01','02','03','04','05','06','07','08']
df = pd.DataFrame(data=data,index=date_index,
    columns=['语文', '数学', '英语'])

# 使用点运算符绘制柱状图
ax = df.plot.bar()

ax.set_title('语数外考试月成绩', fontsize=15)
ax.set_xlabel('考试月',fontsize=12)
ax.set_ylabel('得分',fontsize=12)
````

### 设置绘图样式-多子图绘制

````python
# 创建数据
data = np.random.randint(60, 100, [8, 3])
date_index = ['01','02','03','04','05','06','07','08']
df = pd.DataFrame(data=data,index=date_index,
    columns=['语文', '数学', '英语'])

# 绘制多个子图
ax = df.plot(subplots=True)  # 返回的ax对象是一个数组
ax[0].set_title('语数外考试月成绩', fontsize=15)
ax[2].set_xlabel('考试月',fontsize=12)
ax[0].set_ylabel('得分',fontsize=11)
ax[1].set_ylabel('得分',fontsize=11)
ax[2].set_ylabel('得分',fontsize=11)
````
